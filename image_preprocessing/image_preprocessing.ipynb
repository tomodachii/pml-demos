```python
import pandas as pd
import numpy as np

from glob import glob

import cv2
import matplotlib.pylab as plt
```

# Reading in Images

```python
dog_files = glob('./data/dogs/*.jpg')
cat_files = glob('./data/cats/*.jpg')
```

```python
img_mpl = plt.imread(cat_files[0])
img_cv2 = cv2.imread(cat_files[0])
```

# Display Images

```python
fig, ax = plt.subplots(figsize=(10, 10))
ax.imshow(img_mpl)
ax.axis('off')
plt.show()
```

# Image Channels

## RGB Channels

```python
# Display RBG Channels of our image
fig, axs = plt.subplots(1, 3, figsize=(15, 5))
axs[0].imshow(img_mpl[:,:,0], cmap='Reds')
axs[1].imshow(img_mpl[:,:,1], cmap='Greens')
axs[2].imshow(img_mpl[:,:,2], cmap='Blues')

axs[0].set_title('Red')
axs[1].set_title('Green')
axs[2].set_title('Blue')
plt.show()
```

# Matplotlib vs cv2 Numpy Arrays
* cv2 reads in channels as BGR
* matplot reads in channels as RGB

```python
fix, axs = plt.subplots(1, 2, figsize=(10, 5))
axs[0].imshow(img_cv2)
axs[1].imshow(img_mpl)
axs[0].axis('off')
axs[1].axis('off')
axs[1].set_title('CV Image')
axs[0].set_title('Matplotlib Image')
```

```python
# Converting from BGR to RGB
img_cv2_rgb = cv2.cvtColor(img_cv2, cv2.COLOR_BGR2RGB)
fig, ax = plt.subplots()
ax.imshow(img_cv2_rgb)
plt.show()
```

# Image Manipulation

```python
img = plt.imread(dog_files[0])
fig, ax = plt.subplots(figsize=(8, 8))
ax.imshow(img)
ax.axis('off')
plt.show()
```

```python
img_gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
fig, ax = plt.subplots(figsize=(8, 8))
ax.imshow(img_gray, cmap='Greys')
ax.axis('off')
ax.set_title('Grey Image')
plt.show()
```

# Resizing and Scaling

```python
img.shape
```

```python
img_resized = cv2.resize(img, None, fx=0.25, fy=0.25)
fig, ax = plt.subplots(figsize=(8, 8))
ax.imshow(img_resized)
ax.axis('off')
plt.show()
```

```python
img_resized.shape
```

```python
# Different Size
img_resized = cv2.resize(img, (100, 200))
fig, ax = plt.subplots(figsize=(8, 8))
ax.imshow(img_resized)
ax.axis('off')
plt.show()
```

```python
# Try out different types of interpolation
img_resized = cv2.resize(img, (5000, 5000), interpolation = cv2.INTER_CUBIC)
fig, ax = plt.subplots(figsize=(8, 8))
ax.imshow(img_resized)
ax.axis('off')
plt.show()
```
